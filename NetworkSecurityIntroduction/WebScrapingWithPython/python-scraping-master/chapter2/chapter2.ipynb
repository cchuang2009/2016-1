Chapter 2. Advanced HTML Parsing
---
`bsObj.findAll(tagName, tagAttributes)`: in order to get a list of all of the tags on the
page, rather than just the first.


- `.findAll(tag, attributes, recursive, text, limit, keywords)`
- `.find(tag, attributes, recursive, text, keywords)`


Example
---
- `findAll({"h1","h2","h3","h4","h5","h6"})`: finf out matches tags that contain any one of those attributes.
- `.findAll(text="the prince")`: matches based on the text content of the tags,

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

#html = urlopen("http://www.pythonscraping.com/pages/warandpeace.html")
html = urlopen("http://127.0.0.1:8000/WebScrapPython/2/warandpeace.html")
bsObj = BeautifulSoup(html, "html.parser")
nameList = bsObj.findAll("span", {"class":"green"})
for name in nameList:
    print(name.get_text())
```

```python
html = urlopen("http://127.0.0.1:8000/WebScrapPython/2/warandpeace.html")
bsObj = BeautifulSoup(html, "html.parser")
allText = bsObj.findAll(id="text")
print(allText[0].get_text())
```

```python
html = urlopen("http://127.0.0.1:8000/WebScrapPython/2/page3.htm")
bsObj = BeautifulSoup(html, "html.parser")

for child in bsObj.find("table",{"id":"giftList"}).children:
    print(child)
```

```python
for sibling in bsObj.find("table",{"id":"giftList"}).tr.next_siblings:
    print(sibling) 
```

```python
import re
html = urlopen("http://127.0.0.1:8000/WebScrapPython/2/page3.htm")

bsObj = BeautifulSoup(html, "html.parser")
#images = bsObj.findAll("img", {"src":re.compile("\.\.\/img\/gifts/img.*\.jpg")})
images = bsObj.findAll("img", {"src":re.compile("\.\/page3\_files\/img.*jpg")})
for image in images: 
    print(image["src"])
```

```python
html = urlopen("http://127.0.0.1:8000/WebScrapPython/2/page2.htm")
bsObj = BeautifulSoup(html, "html.parser")

tags = bsObj.findAll(lambda tag: len(tag.attrs) == 2)
for tag in tags:
    print(tag)
```
